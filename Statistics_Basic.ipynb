Question 3: Define mean, median, and mode. Explain why these measures of central
tendency are important.

    Answer:Mean
Definition

The mean (arithmetic average) is the sum of all observations divided by the total number of observations.

    median:
    The median is the middle value of an ordered dataset.

    Mode
Definition

The mode is the value that occurs most frequently in a dataset.


    Importance of Mean, Median, and Mode
    (a) Data Summarization

    Represent large datasets using a single value

    Makes data easy to understand

    (b) Understanding Data Distribution

    Comparing mean and median shows skewness

    Mode shows most common value

    (c) Decision Making

    Mean: average salary, average marks

    Median: income data, house prices

Mode: most popular product size or color

(d) Foundation for Advanced Analysis

Used in variance, standard deviation, correlation, and ML algorithms

Many algorithms assume data centered around the mean

Question 4: Explain skewness and kurtosis. What does a positive skew imply about the
data?

    Answer:Skewness
Definition

Skewness measures the degree of asymmetry of a data distribution around its mean.

In a perfectly symmetrical distribution (normal distribution), skewness = 0.

    Kurtosis
Definition

Kurtosis measures the peakedness and tail heaviness of a distribution compared to a normal distribution.

It indicates the presence of outliers.


    A positive skew (right skew) implies that:

    Majority of data values are lower

    Few extremely high values exist

    Mean is greater than median

    Data is not normally distributed

    Median is often a better measure of central tendency than mean

```python
# Question 5: Implement a Python program to compute the mean, median, and mode of
# a given list of numbers.
# numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]

import statistics

numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26,]
mean = statistics.mean(numbers)
median = statistics.median(numbers)
mode = statistics.mode(numbers)

print("Mean:", mean)
print("Median:", median)
print("Mode:", mode)
```

```python
# Question 6: Compute the covariance and correlation coefficient between the following
# two datasets provided as lists in Python:
# list_x = [10, 20, 30, 40, 50]
# list_y = [15, 25, 35, 45, 60]

import numpy as np

list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

# Covariance
cov_matrix = np.cov(list_x, list_y)   # sample covariance
covariance = cov_matrix[0, 1]

# Correlation
correlation = np.corrcoef(list_x, list_y)[0, 1]

print("Covariance:", covariance)
print("Correlation coefficient:", correlation)
```

```python
# Question 7: Write a Python script to draw a boxplot for the following numeric list and
# identify its outliers. Explain the result:
# data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]


import numpy as np
import matplotlib.pyplot as plt

# Given data
data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

# Create boxplot
plt.figure()
plt.boxplot(data)
plt.title("Boxplot of Given Data")
plt.show()

# Identify outliers using IQR method
q1 = np.percentile(data, 25)
q3 = np.percentile(data, 75)
iqr = q3 - q1

lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

outliers = [x for x in data if x < lower_bound or x > upper_bound]

q1, q3, iqr, lower_bound, upper_bound, outliers
```

```python
# Question 8: You are working as a data analyst in an e-commerce company. The
# marketing team wants to know if there is a relationship between advertising spend and
# daily sales.
# ● Explain how you would use covariance and correlation to explore this
# relationship.
# ● Write Python code to compute the correlation between the two lists:
# advertising_spend = [200, 250, 300, 400, 500]
# daily_sales = [2200, 2450, 2750, 3200, 4000]

import numpy as np

advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]

# Compute correlation coefficient
correlation = np.corrcoef(advertising_spend, daily_sales)[0, 1]

print("Correlation coefficient:", correlation)
```

```python
# Question 9: Your team has collected customer satisfaction survey data on a scale of
# 1-10 and wants to understand its distribution before launching a new product.
# ● Explain which summary statistics and visualizations (e.g. mean, standard
# deviation, histogram) you’d use.
# ● Write Python code to create a histogram using Matplotlib for the survey data:
# survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]

# Answer:Summary Statistics to Use (and Why)

# Before launching a new product, we want to understand overall satisfaction, consistency, and distribution.

# (a) Mean

# Shows the average customer satisfaction

# Helps answer: Are customers generally satisfied or not?

# (b) Median

# Shows the middle score

# Useful if some customers give very low or very high ratings

# (c) Standard Deviation

# Measures variability in opinions

# Low SD → customers agree

# High SD → mixed opinions

# (d) Minimum & Maximum

# Shows the range of satisfaction

# Identifies worst and best experiences

import matplotlib.pyplot as plt

survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]

plt.figure()
plt.hist(survey_scores, bins=7)
plt.xlabel("Survey Scores")
plt.ylabel("Frequency")
plt.title("Histogram of Customer Satisfaction Survey Scores")
plt.show()
```
